```python
cat = Table(fitsio.read('/global/cfs/cdirs/desi/spectro/redux/iron/zcatalog/zpix-main-bright.fits', columns=columns))
```

```python
mask = cat['DESI_TARGET'] & 2**61 > 0  # MWS_ANY
print(np.sum(mask)/len(mask))
```

```python
cat = cat[mask]
```

```python
with warnings.catch_warnings():
    warnings.simplefilter('ignore')
    cat['gmag'] = 22.5 - 2.5*np.log10(cat['FLUX_G']) - 3.214 * cat['EBV']
    cat['rmag'] = 22.5 - 2.5*np.log10(cat['FLUX_R']) - 2.165 * cat['EBV']
    cat['zmag'] = 22.5 - 2.5*np.log10(cat['FLUX_Z']) - 1.211 * cat['EBV']
```

```python
mask = np.isfinite(cat['gmag']) & np.isfinite(cat['rmag']) & np.isfinite(cat['zmag'])
print(np.sum(mask)/len(mask))
cat = cat[mask]
```

```python
cat['EFFTIME_LRG'] = cat['TSNR2_LRG'] * 12.15
plt.hist(cat['EFFTIME_LRG'][(cat['EFFTIME_LRG']<1e3) & (cat['EFFTIME_LRG']>0)], 100);
```

```python
mask = cat['EFFTIME_LRG']>150
mask &= cat['COADD_FIBERSTATUS']==0
print(np.sum(mask)/len(mask))
cat = cat[mask]
```

```python
mask = cat['SPECTYPE']=='STAR'
print(np.sum(mask)/len(mask))
cat = cat[mask]
```

```python
mask = np.full(len(cat), True)
idx = np.random.choice(np.where(mask)[0], size=int(1e5), replace=False)
plt.figure(figsize=(10, 10))
plt.plot((cat['rmag']-cat['zmag'])[idx], (cat['gmag']-cat['rmag'])[idx], '.', ms=0.2)
plt.xlabel('r - z')
plt.ylabel('g - r')
plt.axis([-1, 3., -0.7, 1.8])
plt.grid(alpha=0.5)
plt.show()
```

```python
tmp = Table()
tmp['SUBTYPE'], tmp['count'] = np.unique(cat['SUBTYPE'], return_counts=True)
tmp['frac (%)'] = tmp['count']/len(cat)*100
tmp['frac (%)'].format = '%.1f'
tmp.sort('count', reverse=True)
tmp
```

```python
ms_dict = {'WD':2, 'A':2, 'B':2}
```

```python
mask = np.full(len(cat), True)
idx = np.random.choice(np.where(mask)[0], size=int(2e4), replace=False)

plt.figure(figsize=(10, 10))
for subtype in tmp['SUBTYPE'][:-1]:
    mask = cat['SUBTYPE'][idx]==subtype
    if subtype in ms_dict.keys():
        ms = ms_dict[subtype]
    else:
        ms = 1
    plt.plot((cat['rmag']-cat['zmag'])[idx][mask], (cat['gmag']-cat['rmag'])[idx][mask], '.', ms=ms, label=subtype)
plt.xlabel('r - z')
plt.ylabel('g - r')
plt.axis([-1, 3., -0.7, 1.8])
plt.grid(alpha=0.5)
plt.legend(markerscale=15)
plt.show()
```
