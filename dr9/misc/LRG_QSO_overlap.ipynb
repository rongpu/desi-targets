```python
from __future__ import division, print_function
import sys, os, glob, time, warnings, gc
import numpy as np
import matplotlib.pyplot as plt
from astropy.table import Table, vstack, hstack, join
import fitsio
```

```python
params = {'legend.fontsize': 'x-large',
         'axes.labelsize': 'x-large',
         'axes.titlesize':'x-large',
         'xtick.labelsize':'x-large',
         'ytick.labelsize':'x-large',
         'figure.facecolor':'w'} 
plt.rcParams.update(params)
```

```python
cat0 = Table(fitsio.read('/Volumes/Samsung T5/tmp/dr9.0/0.49.0/dr9_sv1_lrg_south_0.49.0_basic.fits'))
cat = Table(fitsio.read('/Volumes/Samsung T5/tmp/dr9.0/0.49.0/dr9_sv1_lrg_south_0.49.0_photom.fits'))
cat = hstack([cat0, cat], join_type='exact')
print(len(cat))
```

```python
lrg_opt = cat['SV1_DESI_TARGET'] & 2**3 > 0
print(np.sum(lrg_opt), '{:.2f}%'.format(np.sum(lrg_opt)/len(cat)*100))

lrg_ir = cat['SV1_DESI_TARGET'] & 2**4 > 0
print(np.sum(lrg_ir), '{:.2f}%'.format(np.sum(lrg_ir)/len(cat)*100))

lrg_sv = (cat['SV1_DESI_TARGET'] & 2**0>0)
print(np.sum(lrg_sv), '{:.2f}%'.format(np.sum(lrg_sv)/len(cat)*100))
```

```python
sv_qso = (cat['SV1_DESI_TARGET'] & 2**2>0)
print(np.sum(sv_qso & lrg_sv), '{:.2f}%'.format(np.sum(sv_qso & lrg_sv)/np.sum(lrg_sv)*100))
print(np.sum(sv_qso & lrg_opt), '{:.2f}%'.format(np.sum(sv_qso & lrg_opt)/np.sum(lrg_opt)*100))
print(np.sum(sv_qso & lrg_ir), '{:.2f}%'.format(np.sum(sv_qso & lrg_ir)/np.sum(lrg_ir)*100))
```
