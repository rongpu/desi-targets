```python
from __future__ import division, print_function
import sys, os, glob, time, warnings, gc
import numpy as np
import matplotlib.pyplot as plt
from astropy.table import Table, vstack, hstack
import fitsio
# from astropy.io import fits
```

```python
params = {'legend.fontsize': 'large',
         'axes.labelsize': 'large',
         'axes.titlesize':'large',
         'xtick.labelsize':'large',
         'ytick.labelsize':'large',
         'figure.facecolor':'w'} 
plt.rcParams.update(params)
```

```python
target_dir = '/global/cfs/cdirs/desi/target/catalogs/dr8'
output_dir = '/global/cscratch1/sd/rongpu/target/catalogs/dr8.0/0.42.0/ra_dec_only'
```

```python
target_class = 'ELG'

target_path_list = glob.glob(os.path.join(target_dir, '0.42.0/targets/main/resolve/dark/targets-dark-hp-*.fits'))
print(len(target_path_list))

target_bits = {'LRG':0, 'ELG':1, 'QSO':2, 'BGS_ANY':60}
target_bit = target_bits[target_class]
# target_columns = ['MORPHTYPE', 'RA', 'DEC', 'EBV', 'FLUX_G', 'FLUX_R', 'FLUX_Z', 'MW_TRANSMISSION_G', 'MW_TRANSMISSION_R', 'MW_TRANSMISSION_Z', 'NOBS_G', 'NOBS_R', 'NOBS_Z', 'FLUX_W1', 'FLUX_W2', 'MW_TRANSMISSION_W1', 'MW_TRANSMISSION_W2', 'MASKBITS', 'PHOTSYS', 'DESI_TARGET']
target_columns = ['RA', 'DEC', 'DESI_TARGET']

#############################################################################

cat = []
for target_path in target_path_list:
    print(target_path)
    tmp = fitsio.read(target_path, columns=['DESI_TARGET'])
    mask = ((tmp["DESI_TARGET"] & (2**target_bit))!=0)
    idx = np.where(mask)[0]
    if len(idx)==0:
        continue
    print(len(idx)/len(tmp), len(idx), len(tmp))
    cat.append(Table(fitsio.read(target_path, columns=target_columns, rows=idx)))
cat = vstack(cat)
print(len(cat))

cat.write(os.path.join(output_dir, 'dr8_{}_ra_dec_only.fits'.format(target_class)))
```

```python
target_path_list = glob.glob(os.path.join(target_dir, '0.42.0/targets/main/resolve/bright/targets-bright-hp-*.fits'))
print(len(target_path_list))

bgs_bits = {'BGS_FAINT':0, 'BGS_BRIGHT':1}
# target_columns = ['MORPHTYPE', 'RA', 'DEC', 'EBV', 'FLUX_G', 'FLUX_R', 'FLUX_Z', 'MW_TRANSMISSION_G', 'MW_TRANSMISSION_R', 'MW_TRANSMISSION_Z', 'NOBS_G', 'NOBS_R', 'NOBS_Z', 'FLUX_W1', 'FLUX_W2', 'MW_TRANSMISSION_W1', 'MW_TRANSMISSION_W2', 'MASKBITS', 'PHOTSYS', 'DESI_TARGET']
target_columns = ['RA', 'DEC', 'BGS_TARGET']

#############################################################################

cat = []
for target_path in target_path_list:
    print(target_path)
    tmp = fitsio.read(target_path, columns=['BGS_TARGET'])
    mask = ((tmp["BGS_TARGET"] & (2**0))!=0) | ((tmp["BGS_TARGET"] & (2**1))!=0)
    idx = np.where(mask)[0]
    if len(idx)==0:
        continue
    print(len(idx)/len(tmp), len(idx), len(tmp))
    cat.append(Table(fitsio.read(target_path, columns=target_columns, rows=idx)))
cat = vstack(cat)
print(len(cat))

cat.write(os.path.join(output_dir, 'dr8_{}_ra_dec_only.fits'.format('BRIGHT_AND_FAINT')))
```
