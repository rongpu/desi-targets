# ELG

```python
cat = Table(fitsio.read('/Users/rongpu/Documents/Data/dr9_tests/targets/ELG_dr8_south.fits'))

with warnings.catch_warnings():
    warnings.simplefilter("ignore")
    cat['gmag'] = 22.5 - 2.5*np.log10(cat['FLUX_G']/cat['MW_TRANSMISSION_G'])
    cat['rmag'] = 22.5 - 2.5*np.log10(cat['FLUX_R']/cat['MW_TRANSMISSION_R'])
    cat['zmag'] = 22.5 - 2.5*np.log10(cat['FLUX_Z']/cat['MW_TRANSMISSION_Z'])
    cat['w1mag'] = 22.5 - 2.5*np.log10(cat['FLUX_W1']/cat['MW_TRANSMISSION_W1'])
    cat['w2mag'] = 22.5 - 2.5*np.log10(cat['FLUX_W2']/cat['MW_TRANSMISSION_W2'])
```

```python
col_list = ['gmag', 'rmag', 'zmag', 'w1mag', 'w2mag']
xlims = [[20, 24.], [20., 25], [19, 25.], [17, 26], [17, 26]]

for index in range(len(col_list)):
    col = col_list[index]
    v = cat[col]
    mask = np.isfinite(v)
    plt.hist(v, 100, range=xlims[index], density=True)
    plt.xlim(*xlims[index])
    plt.xlabel(col)
    plt.grid(alpha=0.5)
    plt.show()
    print('1st, 5th, 10th, 50th, 90th, 95th, 99th percentiles:')
    print(np.nanpercentile(v, [1, 5, 10, 50, 90, 95, 99]))
```
