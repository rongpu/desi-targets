__Find "error" in the log files__

```python
filelist = glob.glob('/global/cfs/cdirs/desi/users/rongpu/everest/*/*/*/logs/*.log')
print(len(filelist))
```

```python
for fn in filelist:
    with open(fn, 'r') as f:
        lines = list(map(str.strip, f.readlines()))
    error = False
    for line in lines:
        if (line.find('error')!=-1) or (line.find('Error')!=-1) or (line.find('ERROR')!=-1):
            error = True
    if error:
        print(fn, lines[-1])
```
