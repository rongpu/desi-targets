Check if repeatability is a good proxy for VI quality

```python
from __future__ import division, print_function
import sys, os, glob, time, warnings, gc
import numpy as np
import matplotlib.pyplot as plt
from astropy.table import Table, vstack, hstack, join
import fitsio
# from astropy.io import fits
```

```python
params = {'legend.fontsize': 'x-large',
         'axes.labelsize': 'x-large',
         'axes.titlesize':'x-large',
         'xtick.labelsize':'x-large',
         'ytick.labelsize':'x-large',
         'figure.facecolor':'w'} 
plt.rcParams.update(params)
```

```python
tt = Table.read('/global/cfs/cdirs/desi/users/rongpu/spectro/vi/truth_table_LRG_v1.2_tt.fits')
zbest = Table.read('/global/cfs/cdirs/desi/users/rongpu/spectro/vi/truth_table_LRG_v1.2_zbest_single_exp.fits')
print(len(tt), len(zbest))
print(np.all(tt['TARGETID']==zbest['TARGETID']))
```

```python
# Remove stars (based on z)
mask = (tt['best z']<0.001) & (tt['best quality']>=2.5)
tt = tt[~mask]
zbest = zbest[~mask]
print(np.sum(mask), len(tt))

# SN_z cut
mask = zbest['sn_z']<0.5
tt = tt[~mask]
zbest = zbest[~mask]
print(np.sum(mask), len(tt))

# FIBER cut
mask = (zbest['FIBER']>2000) & (zbest['FIBER']<2500)
tt = tt[~mask]
zbest = zbest[~mask]
print(np.sum(mask), len(tt))
```

```python
sv0_lrg = zbest['CMX_TARGET'] & (2**10) > 0
print('SV0_LRG', np.sum(sv0_lrg))
mini_sv_lrg = zbest['CMX_TARGET'] & (2**53) > 0
print('MINI_SV_LRG', np.sum(mini_sv_lrg))
```

```python
outlier_threshold = 0.0033

good_z = (tt['best quality']>=2.5) & (np.abs(zbest['Z']-tt['best z'])<=outlier_threshold)
# good_z = (tt['best quality']>=2.5)
print(np.sum(~good_z))

failures_per_object = np.zeros(len(zbest))
for index in range(len(zbest['Z_SUBSET'][0])):
    z_subset, z_truth = zbest['Z_SUBSET'][:, index], zbest['Z']
    failures_per_object += np.abs((z_subset - z_truth)) > outlier_threshold
not_repeatable = failures_per_object>0
print(np.sum(not_repeatable))
```

```python
plt.hist(failures_per_object);
```

```python
bad_but_repeatable = (~good_z) & (~not_repeatable)
print('not repeatable:', np.sum(not_repeatable), '  bad VI:', np.sum(~good_z))
print('bad but repeatable:', np.sum(bad_but_repeatable), np.sum(bad_but_repeatable)/np.sum(~good_z))
```

```python
plt.figure(figsize=(7, 5))
plt.hist(zbest['zfibermag'], bins=50, alpha=0.3, range=(18, 23), label='All', zorder=1, log=True)
plt.hist(zbest['zfibermag'][~good_z], bins=50, alpha=0.5, range=(18, 23), label='VI quality < 2.5', zorder=3)
plt.hist(zbest['zfibermag'][not_repeatable], bins=50, alpha=0.5, range=(18, 23), label='not repeatable', zorder=2)
plt.xlabel('zfibermag')
plt.title('SV0_LRG')
plt.legend()
plt.grid(alpha=0.5)
plt.show()
```

```python
plt.figure(figsize=(7, 5))
plt.hist(zbest['DELTACHI2'], bins=50, alpha=0.3, range=(0, 1000), label='All', zorder=1, log=True)
plt.hist(zbest['DELTACHI2'][~good_z], bins=50, alpha=0.5, range=(0, 1000), label='VI quality < 2.5', zorder=3)
plt.hist(zbest['DELTACHI2'][not_repeatable], bins=50, alpha=0.5, range=(0, 1000), label='not repeatable', zorder=2)
plt.xlabel('DELTACHI2')
plt.title('SV0_LRG')
plt.legend()
plt.grid(alpha=0.5)
plt.show()

plt.figure(figsize=(7, 5))
plt.hist(np.log10(zbest['DELTACHI2']), bins=50, alpha=0.3, range=(0, 4), label='All', zorder=1, log=True)
plt.hist(np.log10(zbest['DELTACHI2'])[~good_z], bins=50, alpha=0.5, range=(0, 4), label='VI quality < 2.5', zorder=3)
plt.hist(np.log10(zbest['DELTACHI2'])[not_repeatable], bins=50, alpha=0.5, range=(0, 4), label='not repeatable', zorder=2)
plt.xlabel('log10(DELTACHI2)')
plt.title('SV0_LRG')
plt.legend()
plt.grid(alpha=0.5)
plt.show()
```

```python
plt.figure(figsize=(9, 7))
plt.plot(zbest['zfibermag'], zbest['DELTACHI2'], '.', color='0.6', ms=3, alpha=0.5, label='SV LRG')
plt.plot(zbest['zfibermag'][~good_z], zbest['DELTACHI2'][~good_z], 'r.', ms=5.5, alpha=1, label='VI bad')
plt.plot(zbest['zfibermag'][bad_but_repeatable], zbest['DELTACHI2'][bad_but_repeatable], 'rx', ms=10, alpha=1, label='VI bad but repeatble', zorder=2)
plt.plot(zbest['zfibermag'][not_repeatable], zbest['DELTACHI2'][not_repeatable], 'bx', ms=6, alpha=1, label='not repeatble', zorder=1)
plt.axhline(9., ls='--', color='0.6', label='ZWARN threshold')
plt.yscale('log')
plt.xlabel('zfiber')
plt.ylabel('DELTACHI2')
plt.axis([18.5, 23., .1, 100000])
plt.legend(markerscale=2)
plt.grid(alpha=0.5)
plt.show()
```

```python
plt.figure(figsize=(9, 7))
plt.plot(zbest['Z'], zbest['DELTACHI2'], '.', color='0.6', ms=3, alpha=0.5, label='SV LRG')
plt.plot(zbest['Z'][~good_z], zbest['DELTACHI2'][~good_z], 'r.', ms=5.5, alpha=1, label='VI bad')
plt.plot(zbest['Z'][bad_but_repeatable], zbest['DELTACHI2'][bad_but_repeatable], 'rx', ms=10, alpha=1, label='VI bad but repeatble', zorder=2)
plt.plot(zbest['Z'][not_repeatable], zbest['DELTACHI2'][not_repeatable], 'bx', ms=6, alpha=1, label='not repeatble', zorder=1)
plt.axhline(9., ls='--', color='0.6', label='ZWARN threshold')
plt.yscale('log')
plt.xlabel('Redrock redshift')
plt.ylabel('DELTACHI2')
plt.axis([-0.1, 2, .1, 100000])
plt.legend(markerscale=2)
plt.grid(alpha=0.5)
plt.show()
```
